## (a) Dataset Description

IMDB ID - The Internet Movie Database (IMDb) is an online database containing information and statistics about movies, TV shows and video games as well as actors, directors and other film industry professionals. Imdb Id uses unique identifiers for each of the entities referenced in IMDb data. 
Title - Title of the movie
Release Year - Year in which the movies were released
Release Date - Date on which the movie was released
Genre - The film genre is a stylistic or thematic category for motion pictures based on similarities either in the narrative elements, aesthetic approach, or the emotional response to the film.
Writers - A screenplay writer (also called screenwriter, scriptwriter, scribe or scenarist) is a writer who practices the craft of screenwriting, writing screenplays. There may be various writers; Story, Dialogue and script writers involved in a movie.
Actors - Referring to the lead actors who play significant roles in movies.
Directors -Directors are the creative leads of the film. The director has a key role in choosing the cast members, production design and all the creative aspects of filmmaking. 
Sequel  - Sequels are movies made after an original movie and take place after that original movie. 
Hit - Industry Hit is considered when a film crosses certain box office collections. 
Flop - If a film released in theaters fails to break even by a large amount, it is considered a box office flop, thus losing money for the distributor, studio, and/or production company that invested in it.

## (xvi) Query 16 : Plot all the popular actors according to their respective movie count in year 2001 to 2012.

```python
movies2001to2012=movies_df[movies_df['releaseYear']<=2012]
movies2001to2012
```

```python
plt.figure(figsize=(10,10))
popular_actor={}
for i in movies2001to2012['actors']:
    j=i.split(' | ')
    for k in j:
        if k in (popular_actor):
            popular_actor[k]+=1
        else:
            popular_actor[k]=1
top_popular_actor={}
for i in popular_actor:
    if popular_actor[i]>15:
        top_popular_actor[i]=popular_actor[i]
df = pd.DataFrame({'Actor':top_popular_actor.keys(), 'MovieCount':top_popular_actor.values()})
am=sns.barplot(x='Actor',y='MovieCount',data=df)
am.set_xlabel('Actor')
am.set_ylabel('MovieCount')
plt.xticks(rotation=90)
plt.tight_layout()
am.set_title("Popular Actors by movie count from years 2001 to 2012",size=20)
```

## (xix) Query 19 : Find out the least and the most popular Genres according to the number of Movies released.

```python
mingenre = min(genre_d.values())
resmin = [key for key in genre_d if genre_d[key] == mingenre]
print("Least Made Genre:",resmin)
maxgenre = max(genre_d.values())
resmax = [key for key in genre_d if genre_d[key] == maxgenre]
print("Most Made Genre:",resmax)
```
